```python
from bs4 import BeautifulSoup
import requests

# 網頁 URL
url = "https://tw.news.yahoo.com/%E6%97%A5%E6%9C%AC%E8%A1%97%E9%81%93%E7%82%BA%E4%BD%95%E5%A6%82%E6%AD%A4%E4%B9%BE%E6%B7%A8%E6%95%B4%E6%BD%94-%E4%BA%94%E5%A4%A7%E7%90%86%E7%94%B1%E8%AE%93%E6%97%A5%E6%9C%AC%E4%BA%BA%E4%BE%86%E5%91%8A%E8%A8%B4%E4%BD%A0-073000221.html"

# 發送請求並解析 HTML
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# 提取標題，假設標題在 <h1> 標籤中
title = soup.find('h1')  # 如果標題是 <h1> 標籤，您可以根據需要修改這裡的標籤
if title:
    title_text = title.get_text()  # 獲取標題文字
else:
    title_text = "未找到標題"

# 提取文章內容，假設內容在 <p> 標籤中
paragraphs = soup.find_all('p')

# 輸出標題和文章內容
print(f"標題: {title_text}")
print("文章內容:")
for para in paragraphs:
    print(para.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

# 請注意這裡使用 'over18' 來處理成人驗證
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
web = requests.get(url, cookies={'over18': '0'})  # 加入 cookies 來繞過 18 歲驗證

# 解析網頁
soup = BeautifulSoup(web.text, "html.parser")

# 找到所有 class 為 'title' 的 div 元素
titles = soup.find_all('div', class_='title')

# 迭代每個 title 元素
for i in titles:
    # 判斷是否有 <a> 標籤
    if i.find('a') is not None:
        # 取得標題文字
        print(i.find('a').get_text())
        
        # 取得 <a> 標籤的 'href' 屬性並拼接成完整 URL
        print(url + i.find('a')['href'], end='\n\n')
```

```python
import requests
from bs4 import BeautifulSoup
import os

# 定义并创建保存图片的目录
save_dir = r'C:\Users\Huwalli\OneDrive\文件\GitHub\Latia-113-1\1121_image'

# 检查目录是否存在，如果不存在则创建它
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# 获取网页内容
web = requests.get('https://www.ptt.cc/bbs/Beauty/M.1732158962.A.BE5.html', cookies={'over18': '1'})
soup = BeautifulSoup(web.text, "html.parser")

# 找到所有 img 标签
imgs = soup.find_all('img')

# 设定图片编号
name = 0

for i in imgs:
    # 输出图片的 src 属性
    if 'src' in i.attrs:  # 确保 img 标签有 src 属性
        img_url = i['src']
        print(img_url)
        
        # 检查 URL 是否为完整的 URL，如果不是，补全它
        if img_url.startswith('http'):
            img_full_url = img_url
        else:
            img_full_url = 'https:' + img_url  # 处理相对路径
        
        # 使用 requests 获取图片内容
        jpg = requests.get(img_full_url)
        
        # 保存图片到本地指定目录
        with open(f'{save_dir}\\test_{name}.jpg', 'wb') as f:
            f.write(jpg.content)  # 写入图片内容
        
        # 图片编号增加
        name += 1
```

```python
import requests
from bs4 import BeautifulSoup

# 設定網頁 URL
url = 'https://invoice.etax.nat.gov.tw/index.html'

# 發送 GET 請求獲取網頁內容
web = requests.get(url)

# 設定編碼為 UTF-8，以避免亂碼
web.encoding = 'utf-8'

# 使用 BeautifulSoup 解析 HTML 內容
soup = BeautifulSoup(web.text, "html.parser")

# 提取中獎號碼所在的元素，這裡選擇 `.container-fluid` 和 `.etw-tbiggest`
td = soup.select('.container-fluid')[0].select('.etw-tbiggest')

# 提取特別獎
ns = td[0].getText().strip()

# 提取特獎
n1 = td[1].getText().strip()

# 提取頭獎，處理換行符號並取得最後 8 位數字
n2 = [td[2].getText().strip()[-8:], td[3].getText().strip()[-8:], td[4].getText().strip()[-8:]]

# 打印結果
print("特別獎:", ns)
print("特獎:", n1)
print("頭獎:", n2)
```

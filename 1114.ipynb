```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

# 建立日期索引和 DataFrame，包含 100 筆資料
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.rand(100, 4), index=dates, columns=list("ABCD"))

# 創建顏色設定，每個圖的分段顏色不同
colors_A = plt.cm.Reds(np.linspace(0.3, 1, len(df)))
colors_B = plt.cm.Blues(np.linspace(0.3, 1, len(df)))
colors_C = plt.cm.Greens(np.linspace(0.3, 1, len(df)))
colors_D = plt.cm.Purples(np.linspace(0.3, 1, len(df)))
color_maps = [colors_A, colors_B, colors_C, colors_D]

# 創建 2x2 子圖布局
fig, axes = plt.subplots(2, 2, figsize=(14, 14))

# 為每個欄位繪製圓餅圖
for i, col in enumerate(df.columns):
    ax = axes[i // 2, i % 2]  # 定位子圖
    ax.pie(df[col], labels=df[col].round(2), colors=color_maps[i], 
           autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
    ax.set_title(f'{col} 欄位圓餅圖', fontsize=14)

# 調整圖表布局並顯示
plt.tight_layout()
plt.show()
```

```python
# 設定 4 個不同的顏色
colors = ['r', 'b', 'g', 'y']

# 創建 4 個子圖，分別顯示 A, B, C, D 的 3D 條形圖
fig, axes = plt.subplots(2, 2, figsize=(14, 14), subplot_kw={'projection': '3d'})

# 為每個欄位繪製 3D 條形圖
for i, col in enumerate(df.columns):
    ax = axes[i // 2, i % 2]  # 定位到子圖
    x = np.arange(len(df))  # x 座標使用資料的行數
    y = np.zeros(len(df))   # 所有的 y 座標設定為 0
    z = np.zeros(len(df))   # 所有的 z 座標設定為 0
    dz = df[col].values     # 高度用欄位數據表示

    ax.bar3d(x, y, z, dx=1, dy=1, dz=dz, color=colors[i], zsort='average')
    ax.set_xlabel('Index')
    ax.set_ylabel('Y')
    ax.set_zlabel(f'{col} Values')
    ax.set_title(f'3D Bar Plot for Column {col}')

# 顯示圖表
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy import stats
# 假設df是您先前生成的DataFrame，並且您需要使用A、B、C、D這四個欄位資料進行檢定
group1 = df['A'].dropna()  # 提取A欄資料並去除空值
group2 = df['B'].dropna()  # 提取B欄資料並去除空值
group3 = df['C'].dropna()  # 提取C欄資料並去除空值
group4 = df['D'].dropna()  # 提取D欄資料並去除空值

# 使用 scipy.stats.f_oneway 進行單因子 ANOVA 檢定
f_statistic, p_value = stats.f_oneway(group1, group2, group3, group4)

# 輸出結果
print("F 統計量:", f_statistic)
print("p-value:", p_value)

# 解釋結果
alpha = 0.05  # 顯著性水平
if p_value < alpha:
    print("拒絕原假設：不同組別之間的均值存在顯著差異。")
else:
    print("未能拒絕原假設：不同組別之間的均值沒有顯著差異。")
```

```python
import numpy as np
import pandas as pd

# 建立日期索引和 DataFrame，包含 100 筆資料
dates = pd.date_range("20130101", periods=100)  # 100 筆資料
df = pd.DataFrame(np.random.rand(100, 4), index=dates, columns=list("ABCD"))

# 隨機選擇 A、B、C 這些欄位名稱填入 E 欄
np.random.seed(0)  # 設置隨機種子以便重現結果
df['E'] = [np.random.choice(['A', 'B', 'C']) for _ in range(len(df))]

# 現在 df 包含了 E 欄，且 E 欄隨機填入 'A', 'B', 或 'C'
df
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 創建數據集（假設您已經有100筆資料）
data = {
    'Score': df['A'] + df['B'] + df['C'],  # 假設用 A、B、C 欄的數值加總作為成績
    'Method': df['E']  # 使用 E 欄作為組別（Method）
}

# 將數據轉換為 DataFrame
df_anova = pd.DataFrame(data)

# 使用 statsmodels 中的 ols 方法進行單因子 ANOVA 分析
model = ols('Score ~ C(Method)', data=df_anova).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# 輸出 ANOVA 表
print(anova_table)
```
